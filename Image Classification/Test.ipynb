```python
import numpy as np
```

```python
def jpeg_to_python_data(path):
    with open(path, 'rb') as f:
        data = f.read()
    int_data = [byte for byte in data]
    return int_data
```

```python
jpeg_file_path = 'train/apples/img_p1_10.jpeg'
image_data = jpeg_to_python_data(jpeg_file_path)
image_data
```
